```python
import re
def replace (file_name, pattern, subst) :
    file_handel = open (file_name, 'r')
    file_string = file_handel.read ()
    file_handel.close ()
    file_string = ( re.sub (pattern, subst, file_string) )
    file_handel = open (file_name, 'w')
    file_handel.write (file_string)
    file_handel.close ()
```

```python
import pandas as pd
from biopandas.pdb import PandasPdb
def ligand_center(name):
    ppdb = PandasPdb().read_pdb(name)
    xc=ppdb.df['ATOM'][ppdb.df['ATOM']['chain_id']=='P']['x_coord']
    yc=ppdb.df['ATOM'][ppdb.df['ATOM']['chain_id']=='P']['y_coord']
    zc=ppdb.df['ATOM'][ppdb.df['ATOM']['chain_id']=='P']['z_coord']
    xmean=np.mean(np.array(xc))
    ymean=np.mean(np.array(yc))
    zmean=np.mean(np.array(zc))
    return xmean, ymean, zmean
```
