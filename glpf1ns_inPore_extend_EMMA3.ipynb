```python
import glob
import numpy as np
import mdtraj.io as mdio
from msmbuilder.io import save_meta,load_meta
def label(z):
    rangeBin=np.arange(20,70.25,0.25)
    if z>70:
        lab=len(rangeBin)
    else:
        for i in range(len(rangeBin)):
            if z<=rangeBin[i]:
                lab=i
                break
    return lab
allLabel2=[]
files=glob.glob('traj/*.cs')
wf=open('filename2.cs','w')
for ff in files:
    wf.write(str(ff)+'\n')
    data=np.loadtxt(ff)
    dataz=data[:,1]
    lab=[]
    for z in dataz:
        lab.append(label(z))
    lab2=np.array(lab)
    allLabel2.append(lab2)
wf.close()
```
