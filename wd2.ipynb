```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager
import numpy as np
import glob
font_path = '/home/disk2/wdd/msm/calibribold.ttf'
font_prop = font_manager.FontProperties(fname=font_path, size=20)
leg_prop = font_manager.FontProperties(fname=font_path, size=16)
```

```python
#files=glob.glob('p*.txt')
#na='p'
#files=[na+'-1.cs',na+'-2.cs']
fig = plt.figure(figsize=(5,3.8))
sub = fig.add_subplot(1,1,1)
lab=['488 nm','488 nm','532 nm', '532 nm', '555 nm', '555 nm']
for i in range(1,7):
    ff='s'+str(i)+'.txt'
#for ff in files:
    print ff
    data=np.loadtxt(ff)
    plt.plot(data[:,0], data[:,1],lw=2.8,label=lab[i-1],alpha=0.8)
    for label in (sub.get_xticklabels() + sub.get_yticklabels()):
        label.set_fontproperties(font_prop)
        label.set_fontsize(16)
    plt.xlabel(r'nm',fontproperties=font_prop)
    plt.ylabel(r'mAU',fontproperties=font_prop)
    leg=plt.legend(loc=0, labelspacing=0.1, prop=leg_prop, scatterpoints=1, markerscale=1, numpoints=1,handlelength=1.5)
    leg.get_frame().set_linewidth(0.0)
    leg.get_frame().set_alpha(0.1)
plt.xlim(350,720)
plt.axvspan(488-0.9, 488+0.9, facecolor='black',alpha=0.5)
plt.axvspan(555-0.9, 555+0.9, facecolor='black',alpha=0.5)
#plt.ylim(200,500)
#sub.set_xticks(np.linspace(10,22,13))    
    #plt.ylim(1,1000)
figname='s.png'
plt.savefig(figname,dpi=600,bbox_inches='tight')
plt.show() 
```
